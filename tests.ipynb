Цвет линии, стили, ограничения

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

fig, ax = plt.subplots()

ax.plot(x, y)

ax.plot(x, np.sin(x), color=(1.0,0.2, 0.3), linestyle='-')

ax.plot(x, x + 5, color='blue', linestyle='--')

ax.plot(x, x + 3, color='g', linestyle=':')

ax.plot(x, np.cos(x), color='0.75', linestyle='-.')

ax.plot(x, x, color='#FFDD44', linestyle='--')
plt.show()
```

```python
fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

ax.plot(x, y, marker='o')

ax.plot(x + 1, y, marker='^')

ax.plot(2 * x, y, marker='*')
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
fig.set_size_inches(12, 10)

x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

ax.plot(x, y, color='red', linestyle='-.', marker='^', alpha=0.3)

ax.plot(x * 2, y, '--g', linewidth=2.5)

ax.set_xticks(np.arange(0, 6, 1))
ax.set_yticks(np.arange(0, 8, 0.5))

ax.grid(color='yellow', linewidth=2, linestyle='-' )

ax.set_xlim(0, 6)
ax.set_ylim(-1, 9)
```

Подписи и метки

```python
fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

ax.plot(x, y, c='red', linestyle='-', marker='o', label='y = x^2')
ax.plot(x * 0.5, y, c='green', linestyle='--', marker='^', label='y = 0.5 * x^2')

ax.set_ylim(0, 11)

ax.set_title('Две параболы')

ax.set_xlabel("Ось x")
ax.set_ylabel("Ось y")

ax.legend(loc= 'upper left')
```

Другие виды графиков

```python
fig, ax = plt.subplots()

n = 50

x = np.random.rand(n)
y = np.random.rand(n)

color = np.random.rand(n)
area = (30 * np.random.rand(n))**2

ax.scatter(x, y, s=area, c=color, marker='o')
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,6))

ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, s=area,  c=color, marker='*', cmap='viridis')
```

```python
fig = plt.figure(figsize=(14, 10))
ax = plt.subplot()

rect1 = plt.bar(np.arange(1000), np.arange(1000) * 4, width=0.5, color='lightblue')
rect2 = plt.bar(np.arange(1000), np.arange(1000) * 2, width=0.5, color='blue')



plt.legend((rect1[0], rect2[0]), ('IT', 'Юристы'))


ax.set_ylabel('Зарплата')
ax.set_xlabel('Опыт работы')
```

```python
fig, ax = plt.subplots()

x = np.arange(0, 100, 1)
y = x * 2

ax.fill_between(x, y, color='skyblue', alpha=0.4)
ax.plot(x, y, color='Slateblue', alpha=0.8)
```

```python
fig, ax = plt.subplots(figsize=(8,8))

labels = 'Работа', 'Отдых', 'Семья', 'Хобби', 'Друзья'
user = (10, 2, 5, 2, 3)

ax.pie(user, labels= labels, explode=(0.07, 0,0,0,0), autopct='%1.1f%%', startangle= 130, shadow=True)
```
